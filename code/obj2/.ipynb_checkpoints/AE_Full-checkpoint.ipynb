## Generate representation
> Continous values (with zeros)  
>> **Selected:** Coupled values (adding time) + delta time

```python
for i in range(N):
    if i%1000 == 0:
        print("Va en el ",i)
    time_kepler[i], process_lc[i] = generate_representation(time_kepler[i], process_lc[i])
```

```python
#borrar nans... arreglo variable
coupled_lc = []
delta_time = []
coupled_time = []
lens_lc = []
for i in range(N):
    mask_nan = np.isnan(process_lc[i])
    coupled_lc.append(process_lc[i][~mask_nan])
    
    time = time_kepler[i][~mask_nan]
    time = time- time.min()
    # calculate delta time --> this could be done after padding is done..
    coupled_time.append(time)
    delta_time.append(np.hstack([[0],np.diff(time)]))
    lens_lc.append(np.sum(~mask_nan))
    
coupled_lc = np.asarray(coupled_lc)
coupled_time = np.asarray(coupled_time)
delta_time = np.asarray(delta_time)
delta_time
```

### Process --scale and center
> center?  
> With what value scale? (min, iqr, std)  
> Multiplicar por -1 por los max pooling?
>> **Selected:** normalizar (unit variance) --already median centered 

```python
#cada curva de luz dividirla por la desviación estandar..
coupled_lc_scaled = []
for i in range(coupled_lc.shape[0]):
    std_i = coupled_lc[i].std(keepdims=True)
    coupled_lc_scaled.append( coupled_lc[i]/std_i )
coupled_lc_scaled = np.asarray(coupled_lc_scaled)
coupled_lc_scaled.shape
```

```python
min_vals =[]
std_vals = []
mad_vals = []
for i in range(N):
    min_vals.append( coupled_lc[i].min())
    std_vals.append( coupled_lc[i].std() )
    
    med = np.median(coupled_lc[i])
    mad_vals.append( np.median(np.abs(coupled_lc[i]-med)) )

#mientras mas chico sea el valor más aplificara
scale_val = np.min(min_vals) #min of mins...  #va a achicar los valores menores a ese min
print("Min of mins: ",scale_val)

scale_val = np.median(min_vals) #median of mins...  #parece andar mejor.. amplifica ..
print("Median of mins: ",scale_val)

scale_val = np.median(std_vals) #median of std...  amplifica harto los valores
print("Median of std: ",scale_val)

scale_val = np.median(mad_vals) #median of MAD...  amplifica muucho los valores
print("Median of MAD: ",scale_val)

scale_val = np.median(1.5*np.asarray(mad_vals))
print("Median of MAD*1.5: ",scale_val)
```

### Define Model

```python
from keras import backend as K
from keras.models import Sequential, Model
from keras.layers import Input, Conv1D, Dense, Flatten, MaxPool1D, Reshape, UpSampling1D, Lambda, AveragePooling1D
from keras.layers import GlobalAveragePooling1D,GlobalMaxPool1D, TimeDistributed, GRU,LSTM, RepeatVector, SimpleRNN
from keras.layers import BatchNormalization, Dropout, ZeroPadding1D, ZeroPadding2D, Cropping1D, Cropping2D, Conv2D, Conv2DTranspose, MaxPool2D,UpSampling2D

from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot

D_dim = 14
```

```python
def conv_bloq(it, filters, kernel_s, pool, conv_pool=False, BN=False, drop=0,padding='same',dil_r=1, act='relu'):
    f1 = Conv1D(filters, kernel_s, strides=1, padding=padding, activation=act,dilation_rate=dil_r)(it)
    if BN:
        f1 = BatchNormalization()(f1)
    #f1 = Conv1D(filters, kernel_s, strides=1, padding=padding, activation=act,dilation_rate=dil_r)(f1)
    #if BN:
    #    f1 = BatchNormalization()(f1)
    if pool!= 0 and dil_r ==1:
        if conv_pool:
            f1 = Conv1D(filters, kernel_s, strides=pool, padding='valid')(f1)
            if BN:
                f1 = BatchNormalization()(f1)
        else:
            #f1 = MaxPool1D(pool_size=pool, strides=pool, padding='valid')(f1)
            f1 = AveragePooling1D(pool_size=pool, strides=pool, padding='valid')(f1)
    if drop != 0:
        f1 = Dropout(drop)(f1)
    return f1

def rnn_bloq(it, units, layers=1, bid=False, gru=True, drop=0):
    f1 = it
    for i in range(layers):
        if gru:
            layer_rec = GRU(units, return_sequences=(i < layers - 1))
        else:
            layer_rec = SimpleRNN(units, return_sequences=(i < layers - 1) ) #LSTM(units, return_sequences=(i < layers - 1))
        if bid:
            f1 = Bidirectional(layer_rec)(f1)
        else:
            f1 = layer_rec(f1)
        if drop > 0.0 and i > 0:  # skip these for first layer for symmetry
            f1 = Dropout(drop)(f1)
    return f1
        
def deconv_bloq(it, filters, kernel_s, pool, conv_pool=False, BN=False, drop=0,padding='same',dil_r=1,act='relu'):
    f1 = it
    if pool != 0 and dil_r ==1:
        if conv_pool:
            f1 = Conv2DTranspose(filters, (kernel_s,1), strides=(pool,1), padding=padding)(f1)
            if BN:
                f1 = BatchNormalization()(f1)
        else:
            f1 = Lambda(lambda x: K.expand_dims(x, axis=-2))(f1) #along channel axis-to process with conv2d
            f1 = UpSampling2D((pool,1))(f1)
            f1 = Lambda(lambda x: K.squeeze(x, axis=-2))(f1)
            
    f1 = Conv1D(filters, kernel_s, strides=1, padding=padding, activation=act, dilation_rate=dil_r)(f1)
    if BN:
        f1 = BatchNormalization()(f1)
    #f1 = Conv1D(filters, kernel_s, strides=1, padding=padding, activation=act, dilation_rate=dil_r)(f1)
    #if BN:
    #    f1 = BatchNormalization()(f1)
    if drop != 0:
        f1 = Dropout(drop)(f1)
    return f1


def encoder_model_CNN1D(input_dim, channels, L=1, filters=8,kernel_s =10,
                        pool=5,BN=False,conv_pool=False,drop=0,padding='same',dil_r=1,time=False ): #parametros estructurales
    it = Input(shape=(input_dim,channels))  #fixed length..
    f1 = it
    start_f = filters
    for l in range(L):
        #if l == L-1:
        #    f1 = conv_bloq(f1, filters, kernel_s, pool,BN=BN,conv_pool=conv_pool,drop=drop,padding=padding,dil_r=dil_r, act='linear') 
        #else:
        f1 = conv_bloq(f1, filters, kernel_s, pool,BN=BN,conv_pool=conv_pool, drop=drop,padding=padding,dil_r=dil_r) 
        
        filters = int(filters*2)
        filters = min(128, filters)
        if time:
            filters = min(32, filters)
        if dil_r != 1:
            dil_r = int(dil_r*2)
    return Model(inputs=it, outputs=f1)


def decoder_model_CNN2D(input_dim, channels, T=0, L=1, filters=8,kernel_s=10,
                    pool=5,BN=False,conv_pool=False,drop=0,padding='same',dil_r=1): #parametros estructurales  
    it = Input(shape=input_dim)
    #f1 = Lambda(lambda x: K.expand_dims(x, axis=-2))(it) #along channel axis-to process with conv2d
    f1 = it
    filters = int(filters*2**(L-1))
    for _ in range(L):
        f1 = deconv_bloq(f1, filters,kernel_s, pool,conv_pool=conv_pool,BN=BN,drop=drop,padding=padding,dil_r=dil_r)
        
        filters = int(filters/2)
        if dil_r != 1:
            dil_r = int(dil_r*2) #o al igual que filtros dividir??
    #out_x = Lambda(lambda x: K.squeeze(x, axis=-2))(out_x)
    out_x = Conv1D(channels, kernel_s, strides=1, padding=padding, activation='linear')(f1)
    
    if T != 0:
        T_model = K.int_shape(out_x)[1]
        delta_T = T - T_model
        padd_len = int(np.abs(delta_T/2))
        if np.abs(delta_T) % 2 ==0:
            left_pad = right_pad = padd_len                
        else:
            left_pad = padd_len+1
            right_pad = padd_len
            
        if delta_T > 0:
            out_x = ZeroPadding1D((0, delta_T))(out_x)
        elif delta_T < 0:
            out_x = Cropping1D((left_pad, right_pad))(out_x)

    return Model(inputs=it, outputs=out_x)

#keras.layers.ConvLSTM2D(filters, kernel_size, strides=(1, 1), padding='valid', data_format=None, dilation_rate=(1, 1), activation='tanh', recurrent_activation='hard_sigmoid', use_bias=True, kernel_initializer='glorot_uniform', recurrent_initializer='orthogonal', bias_initializer='zeros', unit_forget_bias=True, kernel_regularizer=None, recurrent_regularizer=None, bias_regularizer=None, activity_regularizer=None, kernel_constraint=None, recurrent_constraint=None, bias_constraint=None, return_sequences=False, go_backwards=False, stateful=False, dropout=0.0, recurrent_dropout=0.0)
```

```python
def mse_masked(y_true, y_pred):
    """ Masked on 0 value.."""
    mask = K.cast(K.not_equal(y_true, 0), K.floatx())
    e = (y_true - y_pred)*mask
    return K.sum(K.square(e), axis=-1)  #dividir por el error??

#https://towardsdatascience.com/deep-quantile-regression-c85481548b5a
#https://arxiv.org/pdf/1808.08798.pdf
def titled_loss(q):
    def loss(y_true, y_pred):
        mask = K.cast(K.not_equal(y_true, 0), K.floatx())
        e = (y_true - y_pred)*mask
        return K.sum( K.maximum(q*e, (q-1)*e), axis=-1)
    return loss

def train_model(model,X,y,epochs=1,batch_size=32, val_split=0.0, val_data=None, initial_epoch=0):
    return model.fit(X,y, epochs=epochs, batch_size=batch_size, validation_split=val_split,validation_data=val_data,
                    initial_epoch=initial_epoch)

#autoencoder.compile(loss=[mse_masked],optimizer='adam')
#autoencoder.compile(loss=[titled_loss(0.1)],optimizer='adam') #0.05
```

```python
#4 capas enves de 5..
encoder = encoder_model_CNN1D(T, 1, L=5, filters=16, pool=2, kernel_s=5, drop=0) #model
encoder.name = "LC_encoder"
encoder_t = encoder_model_CNN1D(T, 1, L=5, filters=16, pool=2, kernel_s=5, drop=0, time=True) #model
encoder_t.name = "T_encoder"
inp_dim = encoder.output_shape[1:] 

it_time = Input(shape=X_time.shape[1:])
it_lc = Input(shape=X_lc_scaled.shape[1:])
encoder_time = encoder_t(it_time)
encoder_lc = encoder(it_lc)
concat_encoder = keras.layers.Concatenate()([encoder_time,encoder_lc]) #or multiply?

f1 = rnn_bloq(concat_encoder, units=inp_dim[1]*2, layers=1, bid=False, gru=True, drop=0)
#f1 = rnn_bloq(concat_encoder, units=inp_dim[1], layers=2, bid=False, gru=True, drop=0)
f1_out = Dense(D_dim, activation='linear')(f1)

encoder_model = Model(inputs=[it_time,it_lc],outputs=f1_out)
print("Parametros modelo: ",encoder_model.count_params())
encoder_model.summary()
SVG(model_to_dot(encoder_model).create(prog='dot', format='svg'))
```

```python
it_embd = Input(shape=(D_dim,))

redim_shape = K.int_shape(concat_encoder)[1:] 
decod_rep = RepeatVector(redim_shape[0])(it_embd)

#f1_decod = conv_bloq(decod_rep, filters=inp_dim[1], kernel_s=8, pool=0, drop=0) # en paper ref no lo hace..
f1_decod = decod_rep

concat_decoder = keras.layers.Concatenate()([encoder_time, f1_decod]) #or multiply?
shape_concat = K.int_shape(concat_decoder)[1:]

#rnn
f1 = GRU(inp_dim[1]*2, return_sequences=True)(concat_decoder) #one layer

decoder = decoder_model_CNN2D(K.int_shape(f1)[1:], 1, L=4, filters=8, pool=3, kernel_s=8,T=T, drop=0) #model
lc_out = decoder(f1)

decoder_model = Model(inputs=[it_time, it_embd], outputs=lc_out)
print("Parametros modelo: ",decoder_model.count_params())
SVG(model_to_dot(decoder_model).create(prog='dot', format='svg'))
```

### evaluate with loss and measure acc on labeled data

```python
def plot_n_lc(real_lc, decoded_lc, n = 10):
    for _ in range(n):
        n_sampled = np.random.randint(0, real_lc.shape[0])
        
        f, (ax1, ax2) = plt.subplots(1, 2, sharex=True,figsize= (15,5))
        ax1.plot(real_lc[n_sampled], 'b.-')
        ax1.set_title("Real")
        ax2.plot(decoded_lc[n_sampled], 'b.-')
        ax2.set_title("Reconstructed")
        plt.show()
        
        print("Zoomed!")
        f, (ax1, ax2) = plt.subplots(1, 2, sharex=True,figsize= (15,5))
        ax1.plot(real_lc[n_sampled][:1000], 'b.-')
        ax1.set_title("Real")
        ax2.plot(decoded_lc[n_sampled][:1000], 'b.-')
        ax2.set_title("Reconstructed")
        plt.show()        

        plt.figure(figsize=(15,5))
        plt.plot(real_lc[n_sampled], 'bo-', label="Real")
        plt.plot(decoded_lc[n_sampled], 'g*-', label="Reconstructed")
        plt.title("Over")
        plt.legend()
        plt.show()
        print("----------------------------------------------------------------")
```
