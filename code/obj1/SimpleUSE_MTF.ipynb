```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np
```

```python
LC = [-0.5*np.random.rand(1000), np.random.normal(0, 1, 1500), -1*np.random.exponential(4, 1440)]
```

```python
plt.plot(LC[1])
plt.show()
```

```python
#pre-process
fluxs = [lc/np.abs(np.min(lc)) for lc in LC]
```

```python
plt.plot(fluxs[1])
plt.show()
```

```python
from code_MTF import det_state_2ways, create_MTF_gN, build_MTF

states_values = det_state_2ways(1,-1, n_sta_up=16, n_sta_low=16) 
states_values
```

```python
transition_m = create_MTF_gN(fluxs[1], n_up=16 , n_down=16, N=2, states_values=states_values)
for i in range(2):
    plt.imshow(transition_m[:,:,i], cmap="RdPu", norm=LogNorm())
    plt.show()
```

```python
#build mtf es la implementacion modo paper
transition_m = build_MTF(fluxs[2], np.linspace(0,1, len(fluxs[2])),
                         n_up=1 , n_down=16, delta_M=0.03125, norm=True)

plt.imshow(transition_m, cmap="RdPu", norm=LogNorm())
plt.show()
```
