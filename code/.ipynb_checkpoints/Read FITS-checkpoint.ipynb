```python
from astropy.io import fits
import matplotlib.pyplot as plt

folder = "../KOI_Data/sample/"
```

```python
import os

fits_available = os.listdir(folder)
fits_available[:3]
```

```python
hdulist = fits.open(folder+fits_available[0])

hdulist.info()
```

```python
hdulist[1].header
```

```python
data = hdulist[1].data

time = data.field('TIME')
lc_init = data.field("LC_INIT")
lc_error = data.field("LC_INIT_ERR")
lc_white = data.field("LC_WHITE")
lc_detrend = data.field("LC_DETREND")
time
```

```python
plt.plot(lc_init[:2000])
plt.title("Quarter stitched initial light cu")
plt.show()
```

```python
plt.plot(lc_error[:2000])
plt.title("Quarter stitched initial light curve ERROR")
plt.show()
```

```python
plt.plot(lc_detrend[:2000])
plt.title("Initial Detrend time series (median)")
plt.show()
```

```python
plt.plot(lc_white[:2000])
plt.title("Initial whitened time series fit")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def detrend_median(values, w=5):
    values = np.asarray(values)
    return_v = values.copy()
    for i in range(values.shape[0]):
        median_filter = np.nanmedian(values[ max(i -(w/2) ,0 ): (i+1) + (w/2) ])
        #print("limite inferior: ",max(i -(w/2) ,0 ))
        #print("Valor central: ",i)
        #print("limite superior: ",(i+1) + (w/2))
        #print(median_filter)
        if not np.isnan(median_filter):
            return_v[i] = values[i] - median_filter
    return return_v
def full_detrend_median(values, w=5):
    f_de = detrend_median(values) #forward detrend
    b_de = detrend_median(f_de[::-1]) #backward detrend
    return b_de[::-1]

our_detrend = detrend_median(lc_init[:2000], 5)
our_detrend2 = full_detrend_median(lc_init[:2000], 5)
plt.plot(our_detrend)
plt.title("Our Detrend")
plt.show()

plt.plot(our_detrend2)
plt.title("Our Detrend v2")
plt.show()

from scipy.signal import medfilt
med_filt = medfilt(lc_init[:2000], kernel_size=5)
plt.plot(lc_init[:2000] - med_filt)
plt.title("Library Detrend")
plt.show()
```
