```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

values = ["RMSE", "MAE", "Autocorr","Diff-M", "Res-corr"]

dic_markers= {}
dic_baseline = {}

dic_rawLC = {"Autocorr": 0.27, "Diff-M": 0.78}

#agrandar fuente de legend?
import matplotlib
font = {'family' : 'normal',
        'size'   : 20}
matplotlib.rc('font', **font)

matplotlib.rc('axes', titlesize=25)     # fontsize of the axes title
matplotlib.rc('axes', labelsize=25)    # fontsize of the x and y labels
```

```python
## baseline defintions
dic_baseline["passband 1-500"] = {values[0]: 1.08, values[1]: 0.62, values[2]: 0.97, values[3]: 0.05, values[4]:0.33}
dic_baseline["passband 50-1500"] = {values[0]: 1.04, values[1]: 0.65, values[2]: 0.83, values[3]: 0.20, values[4]:0.24}
dic_baseline["passband 50-2500"] = {values[0]: 0.96, values[1]: 0.64, values[2]: 0.67, values[3]: 0.36, values[4]:0.04}
dic_baseline["mov avg 3"] = {values[0]: 0.72, values[1]: 0.46, values[2]: 0.70, values[3]: 0.27, values[4]:0.03}
dic_baseline["mov avg 5"] = {values[0]: 0.84, values[1]: 0.51, values[2]: 0.78, values[3]: 0.17, values[4]:0.18}
dic_baseline["mov avg 10"] = {values[0]: 0.94, values[1]: 0.55, values[2]: 0.84, values[3]: 0.09, values[4]:0.26}

dic_markers["passband 1-500"] = "o"
dic_markers["passband 50-1500"] = "o"
dic_markers["passband 50-2500"] = "o"
dic_markers["mov avg 3"] = ">"
dic_markers["mov avg 5"] = ">"
dic_markers["mov avg 10"] = ">"


methods = list(dic_baseline.keys())
methods
```

```python
ejex = "RMSE"
ejey = "Autocorr" #correlacion de curva: smoothness

plt.figure(figsize=(10,10))
for value in methods:
    plt.scatter(dic_baseline[value][ejex], dic_baseline[value][ejey], label=value, s=250, marker=dic_markers[value])

#if ejex in dic_rawLC:
#    plt.vlines(y= dic_rawLC[ejex], xmin=0, xmax=10, label="raw light curve" )
#if ejey in dic_rawLC:
#    plt.hlines(y= dic_rawLC[ejey], xmin=0, xmax=10, label="raw light curve" )
    
plt.xlabel(ejex)
plt.ylabel(ejey)
plt.legend()
plt.show()
```

```python
ejex = "RMSE"
ejey = "Diff-M" #inverso a smoothness

plt.figure(figsize=(10,10))
for value in methods:
    plt.scatter(dic_baseline[value][ejex], dic_baseline[value][ejey], label=value, s=250, marker=dic_markers[value])

    plt.xlabel(ejex)
plt.ylabel(ejey)
plt.legend()
plt.show()
```
