```python
lc_measurements = np.load(folder+"K2_PLANET_lc_detr.npy", allow_pickle=True)
lc_time = np.load(folder+"K2_PLANET_time.npy", allow_pickle=True)
lc_measurements
```

```python
mask_downl = ~np.asarray([len(value)==1 and np.isnan(value[0]) for value in lc_time])

lc_measurements = lc_measurements[mask_downl]
lc_time = lc_time[mask_downl]

len(lc_measurements)
```

```python
## save data that were downloaded

df_aux[mask_downl].to_csv(folder+"/K2_CAND_dataset.csv",index=False)

np.save(folder+"/K2_CAND_time.npy",lc_time)
np.save(folder+"/K2_CAND_lc_detr.npy",lc_measurements)
```
